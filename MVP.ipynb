# Engineering MVP

Connie Xiao

The goal for this project was to create an interactive dashboard with real-time data for COVID-19 on streamlit. Data was taken from John Hopkin's CSSE Github. There were 5 files that were used: deaths, recovery, confirmed cases, delta variant, and countries. All the files contained datapoints for each country for each day starting Jan. 22, 2020 to present day. My number one priority for the dashboard was to make it easy to interpret and analyze. 


Here I displayed the up-to-date death counts, confirimed cases, and recovered cases as well as daily cases.

Then I wanted to show the top 20 countries with the highest number of confirmed cases.


I also want to show the statitics by country via drop-down menu along with line graphs to show the increase number of cases by the month.
